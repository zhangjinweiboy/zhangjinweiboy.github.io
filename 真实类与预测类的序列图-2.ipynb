```python
# 绘制单组柱状图

df = predicted
x = list(df["id"])
y_true = list(df["y_true"])
y1 = list(df["y_RandomForest"])
y2 = list(df["y_tree"])
y3 = list(df["y_naive_bayes"])
bar = (
    Bar(init_opts=opts.InitOpts (theme= ThemeType.WHITE,           # 可通过 theme= ThemeType.WHITE 更换主题
                                 bg_color="white",                 # 修改背景颜色，若删除则变为黑色，不便于观看。
                                 width="1000px", height="500px"    # 可通过此处调整图片大小，也可以删除此处。
                                 )) 
    .add_xaxis(x) 
    .add_yaxis("y_True", y_true, gap="2%") 
    .add_yaxis("y_RandomForest(97%)", y1, gap="2%")  
    .add_yaxis("y_tree(96%)", y2, gap="2%")
    .add_yaxis("y_naive_bayes(91%)", y3, gap="2%")# "2011"在图例中显示，不可删除。 通过 gap="2%" 设置不同柱之间的间距，不需要可以删除。
   # 可以在此处添加标记点和标记线。在括号内添加 markpoint_opts=opts.MarkPointOpts(data = [opts.MarkPointItem(type_ = "max")]),markline_opts=opts.MarkLineOpts(data =[opts.MarkLineItem( y = 12 ) ] )  
    ## 添加全局设置
    .set_global_opts( xaxis_opts=opts.AxisOpts(name = "点序列", axislabel_opts=opts.LabelOpts(rotate=0)), # 调整X轴标签的角度，可表示为-35
                      yaxis_opts=opts.AxisOpts(name = "分类类别"),
                      title_opts=opts.TitleOpts( title ="真实类与预测类的序列图", subtitle="" ),
                      toolbox_opts=opts.ToolboxOpts(),               # 展示工具箱
                      legend_opts=opts.LegendOpts(is_show=True) ,    # 显示图例
                      datazoom_opts=opts.DataZoomOpts(is_show=True, type_="slider"), # 显示缩放条,type_="inside"会隐藏缩放条，但是可以缩放。需要显示缩放条，删除type_="inside"即可。
                      tooltip_opts=opts.TooltipOpts(trigger="axis", axis_pointer_type="cross") # 添加坐标线，选中某个点时，会出现坐标虚线，不需要可以删除。
                      )  
    # .reversal_axis()  # 翻转柱状图
    ## 添加标记点，标记线
    .set_series_opts(
        label_opts=opts.LabelOpts(is_show=False), # 在每个条形图上显示数值
        # markline_opts=opts.MarkLineOpts(data = [opts.MarkLineItem(y=3)] )#type_="average",name = "平均值"
    )
)
# 输出为一个html的文件
# bar.render()
# 在notebook中显示
bar.render("/Users/zhangjinwei/Desktop/真实类与预测类的序列图.html")
bar.render_notebook()
```
